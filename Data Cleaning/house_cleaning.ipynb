```python
import re
import numpy as np
import pandas as pd

def extract_floor_info(s):
    if pd.isna(s):
        return pd.Series([np.nan, np.nan])
    
    s = str(s).lower()

    # --- Floor Number (houses case) ---
    floor_match = re.search(r'floors?\s*:\s*(\d+)', s)
    floor_number = float(floor_match.group(1)) if floor_match else np.nan

    # --- Total Floors (not present in your sample → keep NaN) ---
    total_match = re.search(r'out of\s*(\d+)', s)
    total_floors = float(total_match.group(1)) if total_match else np.nan

    return pd.Series([floor_number, total_floors])


df[['floor_number_feat', 'total_floors_feat']] = df['Features'].apply(extract_floor_info)
```

```python
import re
import numpy as np
import pandas as pd

def price_to_crore(x):
    if pd.isna(x):
        return np.nan
    
    s = str(x).lower().replace(",", "").replace("pkr", "").strip()

    # match number + unit
    m = re.search(r"(\d+(?:\.\d+)?)\s*(crore|lakh|arab|thousand)", s)
    
    if not m:
        return np.nan
    
    val = float(m.group(1))
    unit = m.group(2)

    if unit == "crore":
        return val
    elif unit == "lakh":
        return val / 100
    elif unit == "arab":   # 1 arab = 100 crore
        return val * 100
    elif unit == "thousand":
        return val / 100000  # 1 crore = 100,000 thousand

    return np.nan


df["price_crore"] = df["Price"].apply(price_to_crore)
```

```python
import re
import numpy as np
import pandas as pd

def area_to_marla(x):
    if pd.isna(x):
        return np.nan
    
    s = str(x).lower().strip()

    # extract number + unit
    m = re.search(r"(\d+(?:\.\d+)?)\s*(marla|kanal)", s)
    
    if not m:
        return np.nan
    
    val = float(m.group(1))
    unit = m.group(2)

    # convert to marla
    if unit == "kanal":
        val = val * 20  # 1 kanal = 20 marla

    # handle invalid / zero values
    if val == 0:
        print(val)
        return np.nan

    return val


df["area_marla"] = df["Area"].apply(area_to_marla)
```

```python
import ast
import re
import pandas as pd
import numpy as np

def to_dict(x):
    if pd.isna(x):
        return {}
    if isinstance(x, dict):
        return x
    s = str(x).strip()
    if s in ("", "{}"):
        return {}
    try:
        return ast.literal_eval(s)
    except Exception:
        return {}

rooms_dicts = df["Rooms"].apply(to_dict)

rooms_df = pd.json_normalize(rooms_dicts)

# lowercase + underscore
rooms_df.columns = [
    re.sub(r"[^a-z0-9]+", "_", c.strip().lower()).strip("_")
    for c in rooms_df.columns
]

# numeric values
for c in rooms_df.columns:
    rooms_df[c] = pd.to_numeric(rooms_df[c], errors="coerce")

# add to df
for c in rooms_df.columns:
    df[c] = rooms_df[c]
```

```python
import pandas as pd

# -----------------------------
# LUXURY WEIGHTS (balanced)
# -----------------------------
luxury_weights = {
    "Parking Spaces": 3,
    "Double Glazed Windows": 2,
    "Central Air Conditioning": 4,
    "Central Heating": 3,
    "Flooring": 2,
    "Electricity Backup": 4,
    "Waste Disposal": 1,

    "Broadband Internet Access": 1,
    "Satellite or Cable TV Ready": 1,
    "Intercom": 1,

    # community
    "Community Lawn or Garden": 2,
    "Community Swimming Pool": 4,
    "Community Gym": 3,
    "Community Centre": 2,
    "Kids Play Area": 2,
    "Day Care Centre": 2,
    "Barbeque Area": 2,
    "Mosque": 1,

    # standalone (IMPORTANT - you missed)
    "Swimming Pool": 4,
    "Lawn or Garden": 2,

    # premium
    "Sauna": 4,
    "Jacuzzi": 4,

    # services
    "Maintenance Staff": 3,
    "Security Staff": 4,
    "Facilities for Disabled": 2,

    # generic buckets (low weight)
    "Other Community Facilities": 1,
    "Other Healthcare and Recreation Facilities": 1,
    "Other Business and Communication Facilities": 1,
}

# -----------------------------
# FURNISHING WEIGHTS
# -----------------------------
furnishing_weights = {
    "Furnished": 5,
    "Flooring": 3,
    "Central Air Conditioning": 4,
    "Central Heating": 3,
    "Double Glazed Windows": 2,
    "Electricity Backup": 2,
    "Intercom": 1,
}

# -----------------------------
# LOWERCASE MAPS
# -----------------------------
lux_w = {k.lower(): v for k, v in luxury_weights.items()}
fur_w = {k.lower(): v for k, v in furnishing_weights.items()}

# -----------------------------
# FEATURE PARSER
# -----------------------------
def feature_set(s):
    if pd.isna(s):
        return set()
    parts = [p.strip() for p in str(s).split("|") if p.strip()]
    return {p.split(":", 1)[0].strip().lower() for p in parts}

feat_sets = df["Features"].apply(feature_set)

# -----------------------------
# SCORES
# -----------------------------
df["luxury_score"] = feat_sets.apply(lambda s: sum(lux_w.get(x, 0) for x in s))
df["furnishing_score"] = feat_sets.apply(lambda s: sum(fur_w.get(x, 0) for x in s))

# -----------------------------
# ADD FURNISHED COLUMN IMPACT (IMPORTANT)
# -----------------------------
df["furnishing_score"] += df["Furnished"].fillna(0) * 5
```
