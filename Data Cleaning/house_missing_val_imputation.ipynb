```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = df.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
df['nearby_locations_and_other_facilities'].value_counts()
```

```python
df.drop(columns=['nearby_locations_and_other_facilities'], inplace=True)
```

```python
df['floors'].value_counts()
```

```python
df['floors'].describe()
```

```python
df['floors'].isnull().sum()
```

```python
df['floors'].fillna(2, inplace=True)
```

```python
df['furnished'].value_counts()
```

```python
df['furnished'].isna().sum()
```

```python
df.drop(columns=['furnished'], inplace=True)
```

```python
df.drop(columns=[
    'other_business_and_communication_facilities',
    'other_community_facilities',
    'other_healthcare_and_recreation_facilities'
], inplace=True)
```

```python
df['lawn_or_garden'].value_counts()
```

```python
df['lawn_or_garden'].describe()
```

```python
df.drop(columns=['lawn_or_garden'], inplace=True)
```

```python
df['servant_quarters'].value_counts()
```

```python
df['servant_quarters'].describe()
```

```python
df['has_servant_room'] = (df['servant_quarters'] > 0).astype(int)
```

```python
df.drop(columns=['servant_quarters'], inplace=True)
```

```python
df['has_servant_room'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['kitchens'].value_counts()
```

```python
df['kitchens'].describe()
```

```python
df['kitchens'] = df['kitchens'].where(df['kitchens'] <= 6)
```

```python
df['kitchens'].fillna(2, inplace=True)
```

```python
df['store_rooms'].value_counts()
```

```python
df['store_rooms'].describe()
```

```python
df['store_rooms'] = df['store_rooms'].where(df['store_rooms'] <= 4)
```

```python
df['store_rooms'].fillna(1, inplace=True)
```

```python
df.drop(columns=['total_rooms'], inplace=True)
```

```python
df.isnull().sum()
```
