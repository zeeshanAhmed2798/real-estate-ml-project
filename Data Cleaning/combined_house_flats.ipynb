```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flats=pd.read_csv("flats_missing_val.csv")
houses=pd.read_csv("house_missing_val.csv")
```

```python
flats.columns
```

```python
houses.columns
```

```python
drop_cols_flats = [
    'link',
    'area',                # already have area_sqft
    'price_per_sqft',      # leakage ❌
    'society',             # you decided not to use
]

flats.drop(columns=drop_cols_flats, inplace=True)
```

```python
drop_cols_houses = [
    'link',
    'area',
    'price_per_sqft',      # leakage ❌
    'society',
    'rooms',               # already extracted
    'servant_quarters_flag',
    'store_rooms_flag',
    'kitchens_flag',
]

houses.drop(columns=drop_cols_houses, inplace=True)
```

```python
flats.rename(columns={'floor': 'floors'}, inplace=True)
```

```python
flats.drop(columns=['floors_in_building'], inplace=True)
```

```python
flats['property_type'] = 0   # flat
houses['property_type'] = 1  # house
```

```python
flats.columns
```

```python
houses.columns
```

```python
flats['has_servant_room'] = (flats['servant_quarters'] > 0).astype(int)
flats.drop(columns=['servant_quarters'], inplace=True)
```

```python
set(flats.columns) == set(houses.columns)
```

```python
df = pd.concat([flats, houses], ignore_index=True)
```

```python
df.head()
```

```python
drop_cols = [
    'powder_room',
    'steam_room',
    'study_room',
    'lounge_or_sitting_room',
    'laundry_room',
    'prayer_room',
    'drawing_room',
    'dining_room',
    'other_rooms'
]

df=df.drop(columns=drop_cols)
```
