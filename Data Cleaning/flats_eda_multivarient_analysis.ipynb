```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("flats_data_unianalysis.csv").drop_duplicates()
df.head()
```

```python
df.shape
```

## Price

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df[df["area_sqft"] < 10000],
    x="area_sqft",
    y="price",
    hue="bedrooms"
)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df[df["area_sqft"] < 10000],
    x="area_sqft",
    y="price_per_sqft",
    hue="bedrooms"
)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df[df["area_sqft"] < 10000],
    x="area_sqft",
    y="price",
    hue="agePossession"
)
```

```python
sns.barplot(data=df, x="bedrooms", y="price", estimator=np.median)
```

```python
sns.barplot(data=df, x="agePossession", y="price", estimator=np.median)
plt.xticks(rotation="vertical")
plt.show()
```

```python
sns.barplot(data=df, x="agePossession", y="area_sqft", estimator=np.median)
plt.xticks(rotation="vertical")
plt.show()
```

```python
sns.scatterplot(data=df, x="luxury_score", y="price")
```
