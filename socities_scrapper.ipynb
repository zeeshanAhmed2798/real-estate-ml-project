```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
x_paths=['//*[@id="rc-tabs-0-tab-A - F"]','//*[@id="rc-tabs-0-tab-G - L"]','//*[@id="rc-tabs-0-tab-M - R"]','//*[@id="rc-tabs-0-tab-S - Z"]']
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver=webdriver.Chrome() # to open your  it not close bec it is interactivative id it not finfish executing
# time.sleep(5)
driver.maximize_window()
url="https://www.zameen.com/society_maps/lahore-1879"
driver.get(url)
print(f"tittle of page {driver.title}")
# print(f"Url of page {driver.current_url}")
data_frames={}
k=0
for xpath in x_paths:
    k+=1
    socities_path=driver.find_element(by=By.XPATH,value=xpath) 
    socities_path.send_keys(Keys.ENTER)
    time.sleep(40)
    html_code = driver.page_source
    soup=BeautifulSoup(html_code,"html.parser")
    lis=soup.find_all(class_="layout-grid_layoutGrid__1Kawk")
    import pandas as pd
    data = []   # list to store (name, link)
    for sop in lis:
        socities = sop.find_all('a')
        for i in socities:
            name = i.text.strip()          # society name
            link = i.get('href')           # society link
            data.append([name, link])      # add to list

    # Convert to DataFrame
    df = pd.DataFrame(data, columns=['Society Name', 'Link'])
    data_frames[k]=df


    
    
```
